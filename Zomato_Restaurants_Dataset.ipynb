```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato (1).csv", encoding="latin1")
```

```python
df.head()
df.info()
df.describe()

```

##**1.How many restaurants are present in the dataset?**

```python
df['Restaurant Name'].nunique()
```

```python
df['Restaurant ID'].nunique()
```

##**2. Which cities have the highest number of restaurants?**

```python
city_counts = df['City'].value_counts()
highest_cities = city_counts.head(10)
print(highest_cities)
##Visulization
plt.figure(figsize=(10,6))
highest_cities.plot(kind='bar')
plt.title("cities with highest restaurants")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

##**3.What are the top countries with the most restaurants?**

```python
countries = df['Country Code'].value_counts()
top_countries = countries.head(10)
print(top_countries)
##visulization
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')
plt.title("Top Countries with most restuarent")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

##**4.Which currency is used in most of the restaurants?**

```python
count_currency = df['Currency'].value_counts()
most_currency = count_currency.head(10)
print(most_currency)
##visulization
plt.figure(figsize=(10,6))
most_currency.plot(kind='bar')
plt.title("Most used currrency")
plt.xlabel("Currency")
plt.ylabel("COunt")
plt.show()
```

##**2.Location & Geography Analysis**

##**5.Which city has the highest average rating**

```python
city_rating = df.groupby('City')['Aggregate rating'].mean()
highest_rating = city_rating.sort_values(ascending=False).head(10)
print(highest_rating)
##visulization
plt.figure(figsize=(10,6))
highest_rating.plot(kind='bar')

plt.title("Average Rating by City")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.show()
```

##**6.Which city has the most highly-rated restaurants (rating â‰¥ 4)?**

```python
high_rated = df[df['Aggregate rating'] >=4 ]
city_high_count = high_rated['City'].value_counts()
top_10 = city_high_count.head(10)
print(top_10)
##visulization
plt.figure(figsize=(10,6))
top_10.plot(kind='bar')
plt.title("Most Highly Rated Restaurants by City")
plt.xlabel("city")
plt.ylabel("Count")
plt.show()
```

##**7. What are the top 10 localities with maximum restaurants?**

```python
locality_count = df['Locality'].value_counts()
top_10 = locality_count.head(10)
print(top_10)
##visulization
plt.figure(figsize=(10,6))
top_10.plot(kind='bar')
plt.title("Top 10 Locality with maximum restaurantas")
plt.xlabel("Locality")
plt.ylabel("count")
plt.show()
```
