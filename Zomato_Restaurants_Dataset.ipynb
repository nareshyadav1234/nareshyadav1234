##**1.How many restaurants are present in the dataset?**

##**2. Which cities have the highest number of restaurants?**

```python
city_counts = df['City'].value_counts()
highest_cities = city_counts.head(10)
print(highest_cities)
```

##**3.What are the top countries with the most restaurants?**

```python
countries = df['Country Code'].value_counts()
top_countries = countries.head(10)
print(top_countries)
```

##**4.Which currency is used in most of the restaurants?**

```python
count_currency = df['Currency'].value_counts()
most_currency = count_currency.head(10)
print(most_currency)
```
